# Чтение категорий

```python
import pandas as pd
import numpy as np

leafs = pd.read_csv('leaf_diseases/plant-pathology-2020-fgvc7/train.csv') 
y = np.array(leafs.drop(["image_id"], axis=1).values)
leafs
```

# Чтение изображений

```python
import os
import glob
from PIL import Image

img_dir = "leaf_diseases/plant-pathology-2020-fgvc7/images"
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
images = []
for f1 in files:
    images.append(Image.open(f1))
```

```python
import matplotlib.pyplot as plt

fix,axes = plt.subplots(2,5, figsize=(15,4), subplot_kw={'xticks':(),'yticks':()})
for target, image, ax in zip(y[:10], images[:10], axes.ravel()):
    ax.imshow(image)
    ax.set_title(target)
```
